```python
import qrcode # 파이썬 안에 들어온 것
# img = qrcode.make("<String>") #내가 원하는 String
# img.save("qrcode.png")
import random

def proverb_back_qrMaker(back):
    img = qrcode.make(f"<{back[0]}>")
    return img

proverbs = [
    ["굴러온 돌이","박힌 돌 뺀다"],
    ["긁어","부스럼"],
    ["돼지에","진주 목걸이"],
    ["꼬리가 길면","밟힌다"],
    ["돌다리도","두들겨 보고 건너라"],
    ["마른하늘에","날벼락"]
]

# img = qrcode.make("<https://github.com/totocm00/python_study>")
# img.save("qrcode.png")

questions = []
correct_answers = []
user_inputs = []
qr_images = []

for i in range(2):  # 2문제 출제
    proverb_list = random.choice(proverbs)
    front = proverb_list[0]
    back = proverb_list[1]

    img = proverb_back_qrMaker(back)

    print(f"문제 {i+1}: {front}")
    user_input = input("답은? ")

    # 저장
    questions.append(front)
    correct_answers.append(back)
    user_inputs.append(user_input)
    qr_images.append(img)

# 결과 출력
print("\n=== 결과 정리 ===")
for i in range(2):
    print(f"{i+1}번 문제: {questions[i]}")
    print(f"  입력한 답: {user_inputs[i]}")
    print(f"  정답: {correct_answers[i]}")
    print(f"  QR코드 표시 중...\n")
    # qr_images[i].show()
    qr_images[i].show()
```

```python
import qrcode
import random

def generate_qr_from_text(text):
    return qrcode.make(text)

proverbs = [
    ["굴러온 돌이", "박힌 돌 뺀다"],
    ["긁어", "부스럼"],
    ["돼지에", "진주 목걸이"],
    ["꼬리가 길면", "밟힌다"],
    ["돌다리도", "두들겨 보고 건너라"],
    ["마른하늘에", "날벼락"]
]

questions = []
correct_answers = []
user_inputs = []

while True:
    front, back = random.choice(proverbs)
    print(f"\n문제: {front}")

    hint_used = False
    first_input = input("첫 번째 답: ").strip()

    if first_input == back:
        user_inputs.append(f"첫 번째 답: {first_input}")
    else:
        # 힌트 제공
        hint = back[0]
        print(f"힌트: {hint}...")
        second_input = input("두 번째 답: ").strip()

        if second_input == back:
            user_inputs.append(f"첫 번째 답: {first_input}\n두 번째 답: {second_input}")
        else:
            user_inputs.append(f"첫 번째 답: {first_input}")

    questions.append(front)
    correct_answers.append(back)

    choice = input("계속 하시겠습니까? (y/n): ").strip().lower()
    match choice:
        case "y":
            continue
        case "n":
            break
        case _:
            print("잘못된 입력입니다. 자동 종료합니다.")
            break

# QR 텍스트 생성
combined_text = ""
for i in range(len(questions)):
    combined_text += f"{i+1}번 문제: {questions[i]}\n"
    combined_text += f"{user_inputs[i]}\n"
    combined_text += f"정답: {correct_answers[i]}\n\n"

final_qr = generate_qr_from_text(combined_text)
print("\n=== 결과 QR코드 생성 완료! 스캔해서 확인하세요 ===")
final_qr.show()
```
